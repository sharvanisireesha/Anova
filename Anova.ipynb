```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
```

```python
#enter exam scores for each group
group1 = [85, 86, 88, 75, 78, 94, 98, 79, 71, 80]
group2 = [91, 92, 93, 85, 87, 84, 82, 88, 95, 96]
group3 = [79, 78, 88, 94, 92, 85, 83, 85, 82, 81]
```

```python
#perform one-way ANOVA
_,p_value=stats.f_oneway(group1, group2, group3)
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are failing to reject null hypothesis")
```

```python

# initialize list elements
data = [ [9,7,4],[8,6,3],[7,9,2],[8,7,3],[7,7,4],[8,7,3],[9,6,2]]
  
# Create the pandas DataFrame with column name is provided explicitly
df = pd.DataFrame(data, columns=['Medication1','Medication2','Medication3'], )
  
# print dataframe.
df
```

```python
#perform one-way ANOVA
_,p_value=stats.f_oneway(df["Medication1"], df["Medication2"],df["Medication3"])
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are failing to reject null hypothesis")
```

```python
score =pd.DataFrame({'Gender': ['Boys','Boys','Boys','Girls','Girls','Girls','Boys','Boys','Boys','Girls','Girls','Girls','Boys','Boys','Boys','Girls','Girls','Girls'],
                   'AgeGrp': ['10 Year','10 Year','10 Year','10 Year','10 Year','10 Year','11 Year','11 Year','11 Year','11 Year','11 Year','11 Year','12 Year','12 Year','12 Year','12 Year','12 Year','12 Year'],
                   'Score': [4, 6, 8, 4, 8, 9, 6, 6, 9,7, 10, 13, 8, 9, 13, 12, 14,16]})
```

```python
score
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#perform two-way ANOVA
model = ols('Score ~ C(Gender) + C(AgeGrp) + C(Gender):C(AgeGrp)', data=score).fit()
val =sm.stats.anova_lm(model, typ=2)
```

```python
#create data
df = pd.DataFrame({'water': np.repeat(['daily', 'weekly'], 15),
                   'sun': np.tile(np.repeat(['low', 'med', 'high'], 5), 2),
                   'height': [6, 6, 6, 5, 6, 5, 5, 6, 4, 5,
                              6, 6, 7, 8, 7, 3, 4, 4, 4, 5,
                              4, 4, 4, 4, 4, 5, 6, 6, 7, 8]})

#view first ten rows of data 
df
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#perform two-way ANOVA
model = ols('height ~ C(water) + C(sun) + C(water):C(sun)', data=df).fit()
sm.stats.anova_lm(model, typ=2)
```
